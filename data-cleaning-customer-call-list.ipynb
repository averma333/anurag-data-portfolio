```python
#reading excel file in data frame
df = pd.read_excel("/kaggle/input/customer-call-list2/Customer_Call_List2.xlsx")
df
```

```python
#dropping duplicates rows
df = df.drop_duplicates()
```

```python
#removing space, backslash from First_Name column
df.loc[:,'First_Name']=df['First_Name'].str.strip(" \\")
```

```python
#removing underscore, dots. space, forwardslash from Last_Name column
df.loc[:,'Last_Name'] = df['Last_Name'].str.strip("_.../")
df
```

```python
#cleaning phone number
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('-','')
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('/','')
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('|','')
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('Nan','')
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('N/a','')
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('Na','')
df["Phone_Number"]

```

```python
#cleaning NaN in data frame
df= df.fillna('')
df
```

```python
#Reformatting Phone_Number
df.loc[:,'Phone_Number']=df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df.loc[:,'Phone_Number']=df['Phone_Number'].str.replace('--','')
df
```

```python
#Splitting Address Column
df[['Street_Address','State','Zip']]=df["Address"].str.split(',',expand=True)
#df["Address"].str.split(',',2, expand=True)
df
```

```python
#Cleaning "Paying Customer" column
df.loc[:,'Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df.loc[:,'Paying Customer']=df['Paying Customer'].str.replace('No','N')
df.loc[:,'Paying Customer']=df['Paying Customer'].str.replace('N/a','')
df
```

```python
#Cleaning "Do_Not_Contact" column
df.loc[:,'Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df.loc[:,'Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df
```

```python
#Dropping Not_Useful_Column
df=df.drop(columns='Not_Useful_Column')
df
```

```python
#Handling None/NaN
df=df.fillna('')
df
```

```python
#Excluding rows where 'Do_Not_Contact==Y'
#df.iloc[:,df['Do_Not_Contact']=='Y']
df_filtered=df[df['Do_Not_Contact']!='Y']
df_filtered.reset_index(drop=True)
#df_filtered has final data for calling purpose
```
